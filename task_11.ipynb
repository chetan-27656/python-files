```python
from datetime import date
import ipywidgets as widgets
from IPython.display import display

def calculate_age(name, year, month, day):
    """Calculates age based on name and birth date and returns a result string."""
    try:
        birth_date = date(year, month, day)
        today = date.today()
        age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
        return f"{name}'s age is: {age} years"
    except ValueError:
        return "Please enter valid date values."

# Create ipywidgets for input
name_input = widgets.Text(description="Name:")
year_input = widgets.IntText(description="Year:")
month_input = widgets.IntText(description="Month:")
day_input = widgets.IntText(description="Day:")
calculate_button = widgets.Button(description="Calculate Age")
output_label = widgets.Label()

# Link button click to calculation
def on_button_click(b):
    result = calculate_age(name_input.value, year_input.value, month_input.value, day_input.value)
    output_label.value = result

calculate_button.on_click(on_button_click)

# Arrange widgets
input_widgets = widgets.VBox([name_input, year_input, month_input, day_input])
display(input_widgets, calculate_button, output_label)
```

```python
import tkinter as tk
from datetime import datetime

def update_time():
    current_time = datetime.now().strftime("%H:%M:%S")
    time_label.config(text=current_time)
    time_label.after(1000, update_time)

# Taking user input
style = input("Enter font style: ")      # e.g., calibri, arial
size = int(input("Enter font size: "))   # e.g., 40

# Set up the main window
root = tk.Tk()
root.title("Digital Clock")

font_style = (style, size, "bold")

# Create and pack the time label
time_label = tk.Label(root, font=font_style, anchor="center", background='purple', foreground='white')
time_label.pack(fill=tk.BOTH, expand=True)

# Start updating time
update_time()

# Run the GUI loop
root.mainloop()
```
